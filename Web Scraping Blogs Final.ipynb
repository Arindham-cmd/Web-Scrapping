```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
titlelist= []
brieflist = []
linklist = []
```

```python
url = 'http://vyshnoo.com/blog.aspx'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36"}\

req = requests.get(url,headers =headers )
```

```python
print(req.text)
```

```python
soup = BeautifulSoup(req.content,features = 'lxml')
```

```python
blog = soup.find_all('div',class_ ='col-xs-12 col-sm-6 col-md-6 col-lg-6 blog-right')
```

```python
print(blog)
```

```python
for names in blog:
    title = names.find({'h2':'col-xs-12 col-sm-6 col-md-6 col-lg-6 blog-right'}).text
    brief = names.find({'p':'col-xs-12 col-sm-6 col-md-4 col-lg-4 blog-box'}).text
    link = names.find({'a':'readmoreBnt'})
    titlelist.append(title)
    brieflist.append(brief)
    linklist.append(link['href'])
    print(title,brief,link['href'])
    
```

```python
print(titlelist)
```

```python
print(brieflist)
```

```python
print(linklist)
```

```python
df_title = pd.DataFrame(titlelist, columns =['Blog Title']) 
print(df_title) 
```

```python
df_brief = pd.DataFrame(brieflist, columns =['Brief Writeup']) 
print(df_brief) 
```

```python
df_link = pd.DataFrame(linklist, columns =['URL of Main Blog']) 
print(df_link) 
```

```python
Blog_Data = pd.concat([df_title, df_brief,df_link], axis = 1)
Blog_Data.head()
```
